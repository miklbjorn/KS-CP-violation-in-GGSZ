# Introduction and simple cross checks of the code
This notebook introduces the functionality of the python classes, and makes a range of simple cross checks of the python classes and functions in the repository. 

## The "Default Model(.py)" used to make all fits
The `DefaultModel` is used to make all fits, using the standard yield equations in terms of `F_i`, `c_i` and `s_i`. It does not know anything about CP violation. When instantiated, it takes as argument a path to a file that holds a D->Kshh decay amplitude

```python
%matplotlib inline
import math
from models import DefaultModel
amp_dir = '/data/lhcb/users/bjoern/Analyses/B2DPi/data/various_studies/ks_cpv/amplitudes'
dm = DefaultModel(
    "{}/Belle2018_KS_default.pickle".format(amp_dir), # path to file with KS amplitude
    binning_file="../python/input/KsPiPi_optimal.pickle" # path to file with definition of binning scheme
    ) 
```

It automatically calculates default values for `F_i`, `c_i` and `s_i` as predicted using the KS model, assuming no CPV (these can be updated later):

```python
dm.print_Fi_ci_si()
```

Given a set of input x and y parameters and total B+ and B- yields, passed in a list: `[xm, ym, xp, yp, Nplus, Nminus]` the binned yields can be calculated, as per the `F_i`, `c_i` and `s_i` currently set in the model. First: calculate `x` and `y` for sensible physics inputs:

```python
import numpy as np
np.sum(np.sqrt(dm.Fi*dm.Fi_inv)*dm.ci)
```

```python
import UtilityFunctions as uf
# Standard set of input physics parameters
gamma = uf.deg_to_rad(75)
rB = 0.1
dB = uf.deg_to_rad(130)
# Corresponding cartesion parameters
xm, ym, xp, yp = uf.get_xy([gamma, rB, dB])
print "Input x and y parameters:"
print "xm = {: 2.4f}\nym = {: 2.4f}\nxp = {: 2.4f}\nyp = {: 2.4f}".format(xm, ym, xp, yp)
```

Then one can calculate the corresponding yields easily

```python
yields = dm.predict_yields([xm, ym, xp, yp], Nplus=2000, Nminus=2000) # N total = 4000, corresponds to 2015+16 data
print "Yields:"
uf.print_yields(yields)
```

Finally, given a set of yields, `DefaultModel` can fit the corresponding `x` and `y` parameters. In this case, it can be seen that we 1) get the exact input `x` and `y` back and that 2) the uncertainties are very much comparable to what was obtained in the actual Run 2 LHCb measurement.

```python
result = dm.fit(yields)
uf.print_xy_result(result)
```

While the uncertainty size is very similar to what is seen in the Run 2 LHCb measurement, the correlation matrix is somewhat different. 

### Cross check with CLEO results
Sadly, I cannot seem to find values of $K_i'$ in the CLEO paper - these would be useful when checking the assumptions going into the $K_\text{L}^0$ amplitude. One can however compare the obtained $c_i'$ and $s_i'$ parameters -- this is however not an independent comparison, as these where measured using constraints based on the same assumptions, going into the $K_\text{L}^0$ amplitudes used in this paper (although they have an overall minus sign wrong on the $K_\text{L}^0$ amplitude -- which is of no significance at all to their measurement).

A first, reasonable cross-check of the code, is wether the $\Delta s_i/c_i$ agree with those obtained my CLEO. Another is wether the predicted $c_i(')/s_i(')$ matches the measurement. In order to make these, we calculate $c_i(')/s_i(')$ using a number of $K_L^0$ amplitude realisations.

```python
recalculate_cisi_KL = True
number_of_relisations = 25

cleo_check_dm = DefaultModel(
    "../amplitude_calculations/output/KS_default.pickle", # path to file with KS amplitude
    binning_file="../python/input/KsPiPi_optimal.pickle" # path to file with definition of binning scheme
    ) 

ci_KL = np.zeros((8, number_of_relisations))
si_KL = np.zeros((8, number_of_relisations))

for i in range(1, 1+number_of_relisations):
    print i
    cleo_check_dm_KL = DefaultModel(
        "../amplitude_calculations/output/KL_seed_{}.pickle".format(i), # path to file with KS amplitude
        binning_file="../python/input/KsPiPi_optimal.pickle" # path to file with definition of binning scheme
        ) 
    for j in range(8):
        ci_KL[j, i-1] = cleo_check_dm_KL.ci[j]
        si_KL[j, i-1] = cleo_check_dm_KL.si[j]
```

We can then proceed to see the average $\Delta c_i/s_i$ and compare to what was obtained by CLEO:

```python
delta_ci = ci_KL.transpose() - cleo_check_dm.ci[0:8]
delta_si = si_KL.transpose() - cleo_check_dm.si[0:8]

avg_delta_ci = np.mean(delta_ci, axis=0)
std_delta_ci = np.std(delta_ci, axis=0)
avg_delta_si = np.mean(delta_si, axis=0)
std_delta_si = np.std(delta_si, axis=0)

print "delta ci"
for i in range(8):
    index = 7 - i
    print "{} : {:1.3f} +- {:1.3f}".format(i+1, avg_delta_ci[index], std_delta_ci[index])
    
print "\ndelta si"
for i in range(8):
    index = 7 - i
    print "{} : {:1.3f} +- {:1.3f}".format(i+1, avg_delta_si[index], std_delta_si[index])
```

The above delta values show a reasonable agreement with the CLEO values (they incorporate additional model uncertainty, when comparing to this analysis). One can also compare the predicted values, with those obtained in the measurements

```python
avg_ci_KL = np.mean(ci_KL, axis=1)
std_ci_KL = np.std(ci_KL, axis=1)
avg_si_KL = np.mean(si_KL, axis=1)
std_si_KL = np.std(si_KL, axis=1)

print "ci_KL"
for i in range(8):
    index = 7 - i
    print "{} : {:1.3f} +- {:1.3f}".format(i+1, avg_ci_KL[index], std_ci_KL[index])
    
print "\n si_KL"
for i in range(8):
    index = 7 - i
    print "{} : {:1.3f} +- {:1.3f}".format(i+1, avg_si_KL[index], std_si_KL[index])
```

Again, there is a pretty good agreement with the CLEO paper, certainly within errors.

## The FullModel(.py) used to generate yields including all (or some) effects
In the full toy setup, yields are generated using the `FullModel` class. It simply integrates the amplitudes over the bins, taking the efficiency-profile and potential CPV and material effects into account. Therefore it does not know about `F_i`, `c_i`, or `s_i` at all.

In this simple cross-check, we just use it to calculate yields in the default (no extra CPV) situation and fit them back using `DefaultModel` to show that the results are consistent.

```python
from FullModel import FullModel
from PhysicalParameters import PhysicalParameters

# Generate dataset using a Full model
fm = FullModel(
        "../amplitude_calculations/output/KS_default.pickle", # file with KS amplitude
        "../amplitude_calculations/output/KL_default.pickle", # file with KL amplitude (not actually relevant for this cal as we leave out all KL effects)
        eps = 0, # no neutral kaon CPV, by settings eps = 0
        parameters = PhysicalParameters(False), # turn off mixing when importing material param
        binning_file="../python/input/KsPiPi_optimal.pickle") # path to file with definition of binning scheme) 
physics_param = [uf.deg_to_rad(75.), 0.1, uf.deg_to_rad(130.)]
data = fm.predict_yields(physics_param, 4000) # Yields calculated using FullModel.py
xy = uf.get_xy(physics_param) # calculate x, y, N for these physics param
print "Input x and y parameters:"
print "xm = {: 2.4f}\nym = {: 2.4f}\nxp = {: 2.4f}\nyp = {: 2.4f}".format(xy[0], xy[1], xy[2], xy[3])

# Fit using a DefaultModel
dm = DefaultModel("../amplitude_calculations/output/KS_default.pickle",
                 binning_file="../python/input/KsPiPi_optimal.pickle")
result = dm.fit(data)
uf.print_xy_result(result) # print fit results
```

As seen, the exact x and y corresponding to the input parameters are obtained again. This is a cross-check that `FullModel.py` calculates yields that agree with `DefaultModel.py` (as we already saw that `DefaultModel.py` generates and fits consistently.


## GammaFitter(.py): Obtaining the physics parameters from the fit results
The fitter returns `x` and `y` parameters, whereas the real interest is in biases on gamma. The class `GammaFitter` can determine best-fit (gamma, rB, deltaB) combinations from the fit `result`s, mirroring the procedure in `gammacombo`, where the optimal values are found in a least-squares fit. 

Above, we obtained `result` from fitting yields generated with `gamma=75deg`, `rB=0.1`, and `dB=130deg`. As seen, `GammaFitter.py` obtains the correct physics parameters, based on the fitted `x` and `y` in `result`.

```python
from GammaFitter import GammaFitter
gf = GammaFitter()
param, err, result = gf.fit_gamma_from_xy_result(result)
print "gamma = {: 2.3f} +/- {: 2.3f}".format(param[0], err[0])
print "rB    = {: 2.3f} +/- {: 2.3f}".format(param[1], err[1])
print "dB    = {: 2.3f} +/- {: 2.3f}".format(param[2], err[2])
```

### Cross-check GammaFitter.py with the Run 2 experimental results
We can take the input `x` and `y` parameters from the Run 2 LHCb GGSZ measurement and try to extract gamma, rB, and deltaB using `GammaFitter`:

```python
param, err, res = gf.fit_gamma_from_xy(ui.run2_xy, ui.run2_xy_cov_matrix)
print "gamma = {: 2.3f} +/- {: 2.3f}".format(param[0], err[0])
print "rB    = {: 2.3f} +/- {: 2.3f}".format(param[1], err[1])
print "dB    = {: 2.3f} +/- {: 2.3f}".format(param[2], err[2])
```

## Many toy studies or exact yields?
Here, the differences between 3 different study setups are investigated. When looking at biases, one can calculate the expected yields and
* fit them directly, ie. with no statistical variations. Then simply look at the bias. This is fastest (as one fit onyl is required) but may miss effects due to statistics
* generate yield sets around the average, with realistic stats yield sizes. Then take the mean bias. This will require many toys, but gets all statistics dependent effects right
* generate yield sets around the average, with very high stats yield sizes. This requires only a few toys, and is probably equivalent to the top option.

In this section we try all three approaches and compare results. The studies are made w/o any KS-CPV or material effects, hence the mean biases are expected to come out at zero.

First we do the toy studies (and save the output, so there is no need to recalculate every single time)

```python
from ToyStudy import ToyStudy
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm
import matplotlib.mlab as mlab

# Generate dataset using a Full model
fm = FullModel(
        "../amplitude_calculations/output/KS_default.pickle", # file with KS amplitude
        "../amplitude_calculations/output/KL_default.pickle", # file with KL amplitude (not actually relevant for this cal as we leave out all KL effects)
        eps = 0, # no neutral kaon CPV, by settings eps = 0
        parameters = PhysicalParameters(False), # turn off mixing when importing material param
        binning_file="../python/input/KsPiPi_optimal.pickle") # path to file with definition of binning scheme) 

# Fit using a DefaultModel
dm = DefaultModel("../amplitude_calculations/output/KS_default.pickle",
                 binning_file="../python/input/KsPiPi_optimal.pickle")

# Input physics param
physics_param = [uf.deg_to_rad(75.), 0.1, uf.deg_to_rad(130.)]
xy = uf.get_xy(physics_param) # calculate x, y, N for these physics param

# Case 1: do not vary yields
avg_yields = fm.predict_yields(physics_param, 4000) # Yields calculated using FullModel.py
direct_result = dm.fit(avg_yields)
direct_xy = direct_result.x
direct_physics_param, direct_physics_param_err, direct_physics_param_result = gf.fit_gamma_from_xy_result(direct_result)

bias_xy = np.array(direct_xy)-np.array(xy)
```

```python
# Case 2: normal stats, many toys
rerun_cases_2 = False
rerun_cases_3 = False
N_toys_to_make = 1000
case_2_file = "output/cross_checks_toy_case_2.pickle"
ts2 = ToyStudy(dm, fm, physics_param, N_signal=4000, N_toys=N_toys_to_make)
if rerun_cases_2:
    ts2.run(accept_fails_if_p_more_than=0.2)
    ts2.save_results(case_2_file)
else:
    ts2.load_results(case_2_file)
    ts2.fit_bias_and_pull()

    
# Case 3: high stats, many toys
case_3_file = "output/cross_checks_toy_case_3.pickle"
ts3 = ToyStudy(dm, fm, physics_param, N_signal=5e6, N_toys=N_toys_to_make)
if rerun_cases_3:
    ts3.run(accept_fails_if_p_more_than=0.2)
    ts3.save_results(case_3_file)
else:
    ts3.load_results(case_3_file)
    ts3.fit_bias_and_pull()
    
def compare_plots(var_name, var_result, var_input, var_index):
    plt.figure(figsize=(15,5))
    
    plt.subplot(131)
    bins = np.linspace(-.2, 0.2, 40)
    _, _, c2 = plt.hist(ts2.get_list("{}_val".format(var_name)), bins, alpha=0.5, label='2')
    n3, _, c3 = plt.hist(ts3.get_list("{}_val".format(var_name)), bins, alpha=0.5, label='3')
    c1 = plt.plot([var_result.x[var_index], var_result.x[var_index]],[0,max(n3)], label="1")
    ip = plt.plot([var_input[var_index], var_input[var_index]],[0,max(n3)], label="input", linestyle=":", color="k")
    plt.legend(handles=[ ip[0], c1[0], c2[0], c3[0]])
    plt.title("{} values".format(var_name))
    
    plt.subplot(132)
    bins = np.linspace(0, 0.025, 40)
    _, _, c2 = plt.hist(ts2.get_list("{}_err".format(var_name)), bins, alpha=0.5, label='2')
    _, _, c3 = plt.hist(ts3.get_list("{}_err".format(var_name)), bins, alpha=0.5, label='3')
    err = math.sqrt(var_result.hess_inv[var_index, var_index])
    c1 = plt.plot([err, err],[0,9], label="1")
    plt.legend(handles=[ c1[0], c2[0], c3[0]])
    plt.title("{} uncertainty".format(var_name))
    
    plt.subplot(133)
    # best fit of data
    pulls2=ts2.get_list("{}_pull".format(var_name))
    (mu2, sigma2, mu2e, sigma2e) = ts2.get_gaus("{}_pull".format(var_name))
    pulls3=ts3.get_list("{}_pull".format(var_name))
    (mu3, sigma3, mu3e, sigma3e) = ts3.get_gaus("{}_pull".format(var_name))
    bins = np.linspace(-3., 3., 40)
    n2, _, c2 = plt.hist(pulls2, bins, alpha=0.5, label='2')
    n3, _, c3 = plt.hist(ts3.get_list("{}_pull".format(var_name)), bins, alpha=0.5, label='3')
    y2 = sum(n2)*mlab.normpdf(bins, mu2, sigma2)*6./40
    l2 = plt.plot(bins, y2, 'b--', linewidth=2, label="fit 2 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(mu2, mu2e, sigma2, sigma2e))
    y3 = sum(n3)*mlab.normpdf(bins, mu3, sigma3)*6./40
    l3 = plt.plot(bins, y3, 'r--', linewidth=2, label="fit 3 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(mu3, mu3e, sigma3, sigma3e))
    plt.legend(handles=[  c2[0], c3[0], l2[0], l3[0]])
    plt.title("{} pull".format(var_name))
    
    plt.savefig("plots/cross_checks/base_fit_toy_{}.pdf".format(var_name),  bbox_inches='tight')

compare_plots("xm", direct_result, xy, 0)
compare_plots("ym", direct_result, xy, 1)
compare_plots("xp", direct_result, xy, 2)
compare_plots("yp", direct_result, xy, 3)
    
plt.show()
```

```python
plt.figure(figsize=(15,15))

var_name = "xm"
plt.subplot(221)
# best fit of data
pulls2=ts2.get_list("{}_bias".format(var_name))
(mu2, sigma2, mu2e, sigma2e) = ts2.get_gaus("{}_bias".format(var_name))
pulls3=ts3.get_list("{}_bias".format(var_name))
(mu3, sigma3, mu3e, sigma3e) = ts3.get_gaus("{}_bias".format(var_name))
bins = np.linspace(-0.005, 0.005, 40)
n2, _, c2 = plt.hist(pulls2, bins, alpha=0.5, label='2')
n3, _, c3 = plt.hist(pulls3, bins, alpha=0.5, label='3')
l1 = plt.plot([bias_xy[0], bias_xy[0]], [0, max(n3)], label='1')
y2 = sum(n2)*mlab.normpdf(bins, mu2, sigma2)*0.01/40
l2 = plt.plot(bins, y2, 'b--', linewidth=2, label="fit 2 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu2, 100*mu2e, 100*sigma2, 100*sigma2e))
y3 = sum(n3)*mlab.normpdf(bins, mu3, sigma3)*0.01/40
l3 = plt.plot(bins, y3, 'r--', linewidth=2, label="fit 3 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu3, 100*mu3e, 100*sigma3, 100*sigma3e))
plt.legend(handles=[ l1[0],  c2[0], c3[0], l2[0], l3[0]])
plt.title("{} bias".format(var_name))

var_name = "ym"
plt.subplot(222)
# best fit of data
pulls2=ts2.get_list("{}_bias".format(var_name))
(mu2, sigma2, mu2e, sigma2e) = ts2.get_gaus("{}_bias".format(var_name))
pulls3=ts3.get_list("{}_bias".format(var_name))
(mu3, sigma3, mu3e, sigma3e) = ts3.get_gaus("{}_bias".format(var_name))
bins = np.linspace(-0.005, 0.005, 40)
n2, _, c2 = plt.hist(pulls2, bins, alpha=0.5, label='2')
n3, _, c3 = plt.hist(pulls3, bins, alpha=0.5, label='3')
l1 = plt.plot([bias_xy[1], bias_xy[1]], [0, max(n3)], label='1')
y2 = sum(n2)*mlab.normpdf(bins, mu2, sigma2)*0.01/40
l2 = plt.plot(bins, y2, 'b--', linewidth=2, label="fit 2 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu2, 100*mu2e, 100*sigma2, 100*sigma2e))
y3 = sum(n3)*mlab.normpdf(bins, mu3, sigma3)*0.01/40
l3 = plt.plot(bins, y3, 'r--', linewidth=2, label="fit 3 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu3, 100*mu3e, 100*sigma3, 100*sigma3e))
plt.legend(handles=[ l1[0],  c2[0], c3[0], l2[0], l3[0]])
plt.title("{} bias".format(var_name))

var_name = "xp"
plt.subplot(223)
# best fit of data
pulls2=ts2.get_list("{}_bias".format(var_name))
(mu2, sigma2, mu2e, sigma2e) = ts2.get_gaus("{}_bias".format(var_name))
pulls3=ts3.get_list("{}_bias".format(var_name))
(mu3, sigma3, mu3e, sigma3e) = ts3.get_gaus("{}_bias".format(var_name))
bins = np.linspace(-0.005, 0.005, 40)

n2, _, c2 = plt.hist(pulls2, bins, alpha=0.5, label='2')
n3, _, c3 = plt.hist(pulls3, bins, alpha=0.5, label='3')
l1 = plt.plot([bias_xy[2], bias_xy[2]], [0, max(n3)], label='1')
y2 = sum(n2)*mlab.normpdf(bins, mu2, sigma2)*0.01/40
l2 = plt.plot(bins, y2, 'b--', linewidth=2, label="fit 2 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu2, 100*mu2e, 100*sigma2, 100*sigma2e))
y3 = sum(n3)*mlab.normpdf(bins, mu3, sigma3)*0.01/40
l3 = plt.plot(bins, y3, 'r--', linewidth=2, label="fit 3 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu3, 100*mu3e, 100*sigma3, 100*sigma3e))
plt.legend(handles=[ l1[0], c2[0], c3[0], l2[0], l3[0]])
plt.title("{} bias".format(var_name))

var_name = "yp"
plt.subplot(224)
# best fit of data
pulls2=ts2.get_list("{}_bias".format(var_name))
(mu2, sigma2, mu2e, sigma2e) = ts2.get_gaus("{}_bias".format(var_name))
pulls3=ts3.get_list("{}_bias".format(var_name))
(mu3, sigma3, mu3e, sigma3e) = ts3.get_gaus("{}_bias".format(var_name))
bins = np.linspace(-0.005, 0.005, 40)
n2, _, c2 = plt.hist(pulls2, bins, alpha=0.5, label='2')
n3, _, c3 = plt.hist(pulls3, bins, alpha=0.5, label='3')
l1 = plt.plot([bias_xy[3], bias_xy[3]], [0, max(n3)], label='1')
y2 = sum(n2)*mlab.normpdf(bins, mu2, sigma2)*0.01/40
l2 = plt.plot(bins, y2, 'b--', linewidth=2, label="fit 2 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu2, 100*mu2e, 100*sigma2, 100*sigma2e))
y3 = sum(n3)*mlab.normpdf(bins, mu3, sigma3)*0.01/40
l3 = plt.plot(bins, y3, 'r--', linewidth=2, label="fit 3 ({:1.2f} +/- {:1.2f}, {:1.2f} +/- {:1.2f})".format(100*mu3, 100*mu3e, 100*sigma3, 100*sigma3e))
plt.legend(handles=[ l1[0],  c2[0], c3[0], l2[0], l3[0]])
plt.title("{} bias".format(var_name))

plt.savefig("plots/cross_checks/base_fit_biases.pdf",  bbox_inches='tight')
```
