# Material interaction studies

Some looking-into material interactions

## Formula's for the regeneration amplitude

For momenta above 20 Gev/c, a good formula is (https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.42.13)

$$
\Delta\sigma_{reg}(A, p) = 4\pi\sin[-124.7^\circ]\times 2.23 \frac{A^{0.758}}{p^{0.614}(\text{GeV/}c)}\text{ mb}
=22.3 \frac{A^{0.758}}{p^{0.614}(\text{GeV/}c)}\text{ mb}
$$

For lower momenta (https://journals.aps.org/prd/pdf/10.1103/PhysRevD.84.111501) fit's the regeneration amplitude in terms of $\Delta\sigma(K^-p)\equiv \sigma(K^-p)-\sigma(K^+p)$ using $Kd$ cross-sections ($A=2$) and PDG data, and finds

$$
\Delta\sigma_{reg} = \frac{A^{0.758}}{1+1.25 \exp[-1.814 p_K (\text{GeV/}c)]}\Delta\sigma(K^-p)
$$

This is looked at below, using PDG data, and compared to a few existing measurements.

Super reference with more detail than will ever be necessary: https://ac.els-cdn.com/0370157382900436/1-s2.0-0370157382900436-main.pdf?_tid=50ead0ed-f1c6-40c5-bad0-a594f3b21fcc&acdnat=1549634644_b3453e2acb10a98d4a9cd707c5b8930d

## A look at actual data
Load experimental data on $\sigma (K^\pm \{p,n,d\})$ from (http://pdg.lbl.gov/2017/hadronic-xsections/hadron.html)

```python
import pandas as pd
import numpy as np
df = {}
args = {'sep':"\s+", 
        'skiprows':11, 
        'usecols':[1,4,5,6,7,8,9], 
        'names':['P', 'SIG', 'STA+', 'STA-', 'SYS+', 'SYS-', 'FLAG']}
df["kpp"] = pd.read_csv("../pdg_data/rpp2016-kpp_total.dat", **args)
df["kmp"] = pd.read_csv("../pdg_data/rpp2016-kmp_total.dat", **args)
df["kpn"] = pd.read_csv("../pdg_data/rpp2016-kpn_total.dat", **args)
df["kmn"] = pd.read_csv("../pdg_data/rpp2016-kmn_total.dat", **args)
df["kpd"] = pd.read_csv("../pdg_data/rpp2016-kpdeut_total.dat", **args)
df["kmd"] = pd.read_csv("../pdg_data/rpp2016-kmdeut_total.dat", **args)



print df["kpp"].head()
```

In order to make processing easier, these are averaged out, using the statistical errors as weight

```python
new_df = {}

def wm(x, w):
    return np.average(x, weights=w)
def wme(x, w):
    if len(x)==1:
        return np.sqrt(1/(w))
    wm_x = wm(x, w)
    var_x = np.sum(np.power(x-wm_x,2)*w)
    sum_w = np.sum(w)
    sum_w2 = np.sum(np.power(w,2))
    
    return np.sqrt(var_x*(sum_w2)/(sum_w*sum_w) * sum_w / (sum_w*sum_w-sum_w2))

for k in df.keys():
    df[k]['TOTAL ERR'] = (df[k]['STA+'] + df[k]['STA-'])/2 + df[k]['SIG']*(df[k]['SYS+'] + df[k]['SYS-'])/(200.)
    df[k]['WEIGHT'] = 1/(df[k]['TOTAL ERR']*df[k]['TOTAL ERR'])
    lambda_wm = lambda x: wm(x, df[k].loc[x.index, 'WEIGHT'])
    lambda_wme = lambda x: wme(x, df[k].loc[x.index, 'WEIGHT'])
    
    f = {'SIG': {'MEAN' : lambda_wm, 'ERR': lambda_wme} }
    # Groupby and aggregate with your dictionary:
    new_df[k] = df[k].groupby(['P'], group_keys=False).agg(f)
    new_df[k].columns = list(map('_'.join, new_df[k].columns.values))
    new_df[k]['P'] = new_df[k].index
#     new_df[k].reset_index()
```

Make interpolation functions, in order to be able to calculate differences at arbitrary momenta

```python
from scipy.interpolate import interp1d
def running_mean(xs, x_errs, N):
    avg = np.zeros(len(xs))
    err = np.zeros(len(xs))
    for i, x in enumerate(xs):
        bin_xs = xs[max(0, i-N) : min(len(xs)-1, i+N)]
        bin_es = x_errs[max(0, i-N) : min(len(xs)-1, i+N)]
        bin_ws = 1/np.power(bin_es,2)
        avg[i] = wm(bin_xs, bin_ws)
        err[i] = wme(bin_xs, bin_ws)
    return avg, err

sig_f = {}
for k in df.keys():
    new_df[k]['SIG_SMOOTH'] = running_mean(new_df[k]['SIG_MEAN'].values, new_df[k]['SIG_ERR'].values, 7)[0]
    new_df[k]['SIG_SMOOTH_ERR'] = running_mean(new_df[k]['SIG_MEAN'].values, new_df[k]['SIG_ERR'].values, 7)[1]
    sig_f[k] = interp1d(new_df[k]['P'], new_df[k]['SIG_MEAN'])

    
print new_df["kpp"].head(), '\n'
print 'sig_mean_interp (0.32066):', sig_f['kpp'](0.32066)
print 'sig_mean_interp (0.33):', sig_f['kpp'](0.33)
```

```python
import numpy as np

def dsig_reg_high_p(A, p, A_power=0.758, p_power =-0.614):
    return 22.3*np.power(A, A_power)*np.power(p, p_power)
def dsig_reg_low_p(A, p):
    dsig_m = sig_f['kmp'](p)
    dsig_p = sig_f['kpp'](p)
    return np.power(A, 0.758)/(1+1.25*np.exp(-1.814*p))*(dsig_m - dsig_p)
def dsig_reg_low_p_simple(A, p):
    dsig_m = sig_f['kmp'](p)
    dsig_p = sig_f['kpp'](p)
    return np.power(A, 0.758)/(1+1.25*np.exp(-1.814*p))*dsig_reg_high_p(1, p)
```

```python
def compare_mp(xs_partner, A=1):
    # funtion that makes comparison plots for either neutrons, protons or deuterium
    key_m = 'km{}'.format(xs_partner)
    key_p = 'kp{}'.format(xs_partner)
    ps_p = new_df[key_p]['P']
    ps_m = new_df[key_m]['P']
    common_ps = np.intersect1d(ps_p, ps_m)
    common_sig_p = new_df[key_p][new_df[key_p]['P'].isin(common_ps)]['SIG_MEAN']
    common_sig_m = new_df[key_m][new_df[key_m]['P'].isin(common_ps)]['SIG_MEAN']
    common_sig_p_e = new_df[key_p][new_df[key_p]['P'].isin(common_ps)]['SIG_ERR']
    common_sig_m_e = new_df[key_m][new_df[key_m]['P'].isin(common_ps)]['SIG_ERR']
    
    diff = common_sig_m - common_sig_p
    diff_e = np.sqrt(common_sig_p_e**2 + common_sig_m_e**2)
        
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    
    
    ax[0].errorbar(x=ps_p, y=new_df[key_p]['SIG_MEAN'], yerr=new_df[key_p]["SIG_ERR"], fmt='.', 
                   label='$\sigma(K+{})$'.format(xs_partner), alpha=0.7)
    ax[0].errorbar(x=ps_m, y=new_df[key_m]['SIG_MEAN'], yerr=new_df[key_m]["SIG_ERR"], fmt='.', 
                   label='$\sigma(K-{})$'.format(xs_partner), alpha=0.7, color='r')
    ax[0].legend()
    
    ax[0].set_xscale('log')
    ax[0].set_yscale('log')
    
    ax[1].errorbar(x=common_ps, y=diff, yerr=diff_e, fmt='.', 
                   label='$\Delta\sigma(K{})$'.format(xs_partner), alpha=0.7, color='g')
    ax[1].plot(common_ps, dsig_reg_high_p(A, common_ps), label="high_p")
    ax[1].plot(common_ps, dsig_reg_low_p(A, common_ps), label="low_p", color='r')
    ax[1].plot(common_ps, dsig_reg_low_p_simple(A, common_ps), label="low_p_simple", color='k')
    
    ax[1].legend()
    ax[1].set_xscale('log')
    ax[1].set_yscale('log')
    
    ax[2].errorbar(x=common_ps, y=common_ps*diff, yerr=diff_e, fmt='.', 
                   label='$\Delta\sigma(K{})$'.format(xs_partner), alpha=0.7, color='g')
    ax[2].plot(common_ps, common_ps*dsig_reg_high_p(A, common_ps), label="high_p")
    ax[2].plot(common_ps, common_ps*dsig_reg_low_p(A, common_ps), label="low_p", color='r')
    ax[2].plot(common_ps, common_ps*dsig_reg_low_p_simple(A, common_ps), label="low_p_simple", color='k')
    ax[2].legend()
    ax[2].set_xscale('log')
    ax[2].set_yscale('log')
```

```python
compare_mp('p')
compare_mp('n')
compare_mp('d', 2)
```

```python
N_A = 6.02e+23 
mK = 0.497
conversion = 1e-27
c = 30
phi = 127./180.*np.pi
def dchi(df, A, rho):
    return - 2 * np.pi * rho * conversion * N_A * c/ (mK*A) * df * (np.cos(phi)+1j*np.sin(phi))
 
dlambda = 5.290 + 1j/2*(1/0.089)
print dlambda
A = 28
rho = 0.02*2.6
p = np.linspace(0.5, 30, 100)

plt.plot(p, abs((1/(2*dlambda))*dchi(p*dsig_reg_high_p(A, p)/(4*np.pi*np.sin(phi)), A, rho)), 
         color='b', label='Default')
plt.plot(p, abs((1/(2*dlambda))*dchi(p*dsig_reg_low_p(A, p)/(4*np.pi*np.sin(phi)), A, rho)), 
         linestyle=':', color='b', label='Low P')

plt.legend()
plt.xlabel('$p_K$ [GeV/c]')
plt.ylabel('$\Delta \chi / 2\Delta\lambda $')
plt.savefig('plots/dchi_over_dlambda.pdf')
```

## Absorbtion comparisons

```python
# Compare at LHCb conditions, p = 35

def sig_to_chi(sig, p, A, rho):
    mK = 0.497
    c = 30
    N_A = 6.02e+23 
    N = rho*N_A/A
    arg_f = 127.4/180.*np.pi
    conversion=1e-27
    Im_f = (p)/(4*np.pi)*sig*conversion
    abs_f = Im_f/np.sin(arg_f)
    f = abs_f*np.cos(arg_f) + 1j * abs_f*np.sin(arg_f)
    chi = 2*np.pi*N/mK*f*c # in 1/ns
    return chi

sigma_Al = 373.4*2
sigma_Si = sigma_Al / np.power(27, 0.84) * np.power(28, 0.84)
chi_Si = sig_to_chi(sigma_Si, 35., 28, 2.56)
print "High p:", chi_Si
print sigma_Si
sigma_C = 350+210
sigma_Si = sigma_C / np.power(12, 0.84) * np.power(28, 0.84)

chi_Si = sig_to_chi(sigma_Si, 1., 28, 2.56)
print "Low p:", chi_Si
print sigma_Si
```
